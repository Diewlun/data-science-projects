```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cc_impr_overlap = pd.read_csv('Cross Channel Impression Overlap.csv', index_col=[0])
```

```python
cc_impr_overlap.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
# mask
mask = np.triu(np.ones_like(cc_impr_overlap, dtype='bool'))

# adjust mask and impr_overlap
mask = mask[1:, :-1]
cc_impr_overlap = cc_impr_overlap.iloc[1:,:-1].copy()

# plot heatmap
sns.heatmap(cc_impr_overlap, mask=mask, annot=True, fmt=".2f", cmap='Reds')
plt.yticks(rotation=0)
plt.savefig('Cross Channel Impression Reach Overlap April 2022.png')
plt.show()
```

```python
cc_pub_overlap = pd.read_csv('Cross Channel Publisher Impression Overlap.csv', index_col=[0])
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(cc_pub_overlap, annot=True, fmt=".2f", cmap='Reds', linewidths=1, linecolor='black', clip_on=False)
# yticks
plt.yticks(rotation=0)
plt.show()
```
